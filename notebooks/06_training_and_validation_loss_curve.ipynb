```python
import pandas as pd

loss_df = pd.read_csv("loss_log_day6.csv")
loss_df.head(), loss_df.shape
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(8, 5))
plt.plot(loss_df["train_loss"], label = "Training Loss", linewidth = 2)
plt.plot(loss_df["val_loss"], label = "Validation Loss", linewidth = 2)

plt.xlabel("Epoch")
plt.ylabel("MSE Loss")
plt.title("CRISPR Model Training vs Validation Loss")
plt.legend()
plt.grid(alpha = 0.3)
plt.show()

```
